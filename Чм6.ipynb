```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

x=np.array([-3.,-2.,0.,2.], dtype=float)
y=np.array([-22.,-13.,-7.,23.], dtype=float)
x_zadani =np.array([-4.,-1.5,-1,1.5], dtype=float)

def lagrange_interpolation(x, y, x_test):
    n = len(x)
    p = np.zeros(n)
    for i in range(n):
        p_i = 1
        for j in range(n):
            if i != j:
                p_i *= (x_test - x[j])/(x[i] - x[j])
        p[i] = p_i
    return np.dot(y, p)

for i in x_zadani:
    f_interp = lagrange_interpolation(x, y, i)
    print("Значення функції у точці ",i," = ", f_interp.round(4))



```

```python
xnew=np.linspace(np.min(x),np.max(x),100)
ynew=[lagrange_interpolation(x,y,i) for i in xnew]

plt.plot(x, y, 'o', color='g')
plt.plot(xnew, ynew, color='r')

plt.title(' Графік функції Лагранжа_1')

plt.grid(True)
plt.xlabel('x')

plt.ylabel('y')

plt.show()
```

```python
f = lagrange(x, y)

fig = plt.figure(figsize = (7,5))

plt.plot(xnew, f(xnew), 'y', x, y, 'bo')

plt.title('Графік функції Лагранжа_2')

plt.grid()

plt.xlabel('x')

plt.ylabel('y')

plt.show()
```
